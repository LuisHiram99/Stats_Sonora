```python
coordenadas_df = df[['NOM_LOC', 'NOM_MUN', 'LONGITUD', 'LATITUD']]
```

```python
import re
import pandas as pd

def dms_to_dd(dms):
    # Regex para extraer grados, minutos, segundos y dirección
    regex = r"(\d+)°(\d+)'([\d\.]+)\"?\s*([NSEW])"
    m = re.match(regex, dms.strip())
    if not m:
        return None
    deg, minutes, seconds, direction = m.groups()
    dd = float(deg) + float(minutes)/60 + float(seconds)/3600
    if direction in ['W', 'S']:
        dd *= -1
    return dd


# Convertir columnas
coordenadas_df["LONG_DEC"] = coordenadas_df["LONGITUD"].dropna().apply(dms_to_dd)
coordenadas_df["LAT_DEC"] = coordenadas_df["LATITUD"].dropna().apply(dms_to_dd)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Filtrar solo filas con coordenadas válidas
coordenadas_df_clean = coordenadas_df.dropna(subset=["LONG_DEC", "LAT_DEC"])

# Crear geometrías
geometry = [Point(xy) for xy in zip(coordenadas_df_clean.LONG_DEC, coordenadas_df_clean.LAT_DEC)]
gdf = gpd.GeoDataFrame(coordenadas_df_clean, geometry=geometry, crs="EPSG:4326")
```
