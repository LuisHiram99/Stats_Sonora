## Aplicamos el mismo preprocesamiento para todos los estados

```python
def preprocess_pobreza_sonora(df):
   '''
   Preprocess the pobreza_mexico_estados dataframe
   '''
   df.drop(index=[0,1,2,3,4,27,28,29,30,31,32], inplace=True)

   df.columns = range(df.shape[1])

   df.iloc[4,0] = df.iloc[4,1]  
   df.iloc[5,0] = df.iloc[5,1]
   df.drop(columns=[1,2,3,9,15], inplace=True)
   df.drop(index=[5,7], inplace=True)
   df.iloc[0,0] = 'index'
   df.columns = ['index', '2016_porc', '2018_porc', '2020_porc', '2022_porc', '2024_porc', '2016_num', '2018_num', '2020_num',
      '2022_num', '2024_num', '2016_prom', '2018_prom', '2020_prom', '2022_prom', '2024_prom']
   df.drop(index=[6,14,17,24], inplace=True)
   df.reset_index(drop=True, inplace=True)
```
